$$
f(x_0+h) = f(x_0)+f^{'}(x_0)h + f^{''}(x_0)\frac{h^2}{2}
$$

Qual o valor da aproximação no ponto $x = 0.5$ (considere três casas decimais)?

Imprima, em um só quadro, o gráfico da função $f(x) = tan(x)$ e do polinômio de Taylor obtido. Além disso, mostre graficamente o ponto $x = 0.5$ avaliado no polinômio de Taylor.


a) 0.437<br>
b) 0.643<br>
c) 0.545 (x)<br>

```python
# Solucao

# Valor da aproximação no ponto x = 0.5
x0 = 0.4
h = 0.1

# Aproximação de grau 2
# polinômino de Taylor 2º grau f(x0 + h) para f(x) = tan(x)
fh = np.tan(x0) + (1/np.cos(x0))**2 * h + 2*np.tan(x0)*((1/np.cos(x0))**2) * (h**2/2)

print("Polinômio de Taylor de grau 2 ao redor do ponto 0.4")
print("Valor de aproximação no ponto x = 0.5:", fh)
#print("Valor exato f(0.5) = tan(0.5):", np.tan(0.5))
print("Erro da aproximação no ponto x = 0.5:", np.tan(0.5) - fh)
```

```python
# Aproxamação de grau 1
# polinômino de Taylor 1º grau f(x0 + h) para f(x) = tan(x)
fh1 = np.tan(x0) + (1/np.cos(x0))**2 * h
print("Polinômio de Taylor de grau 1 ao redor do ponto 0.4")
print("Valor de aproximação no ponto x = 0.5:", fh1)
print("Erro da aproximação no ponto x = 0.5:", np.tan(0.5) - fh1)
```

### Questão 3:  (2,5 pontos)
Considere a equação do calor como descrita abaixo:

$$
u_t(x,t)=\kappa u_{xx}(x,t)
$$

Supondo $x\in[0,2\pi]$, $t\in[0,3]$, $\kappa=3$, $\Delta x = 0.7$ e as condições de contorno $u(0,t)=t^3$ e $u(2\pi,t)=sen(t)$ encontre uma aproximação do valor da solução em $u(2,3)$. O valor aproximado é:

a) 8.326<br>
b) 7.975 (x)<br>
c) 6.447<br>
d) 5.450<br>

Faça um gráfico da solução aproximada em $u(x,3)$ e o ponto $u(2,3)$ apontado na solução.

Dica: calcule a solução aproximada em $u(x,3)$ por interplação linear

```python
# Solucao

# Condição de estabilidade solução numérica
# Método: diferenças finitas
dx = 0.7
kappa = 3

print("Passo dt para estabilidade do método numérico (menor igual a):", 0.5*dx**2/kappa)
```

```python
# Solução numérica - Equação Calor
# Método: diferenças finitas

# Parâmetros
dx = 0.7
dt = 0.08
kappa = 3

# Limites espaço - tempo
xmin = 0
xmax = 2*np.pi # 3.141592653589793
tmin = 0
tmax = 3
xpassos = int((xmax - xmin)/dx + 1)
tpassos = int((tmax - tmin)/dt + 1)

x = np.linspace(xmin, xmax, int(xpassos)) # [0, 2pi], dx
t = np.linspace(tmin, tmax, int(tpassos)) # [0, 3], dt

print(x) # dx = 0.78539816
print(t) # dt = 0.08108108
print('Número de passos na discretização espacial:', xpassos)
print('Número de passos na discretização temporal:', tpassos)
```

```python
# Será que dx e dt reais continuam dentro da condição de estabilidade?
dx = x[1]
dt = t[1]
print('dx real:', dx)
print('dt real:', dt)
print("Passo dt para estabilidade do método numérico (menor igual a):", 0.5*dx**2/kappa) # sim, continua estável
```

> Com a condição de estabilidade respeitada, podemos prosseguir para o computo da solução numérica da equação de calor pelo método das diferenças finitas.

```python
u = np.zeros((xpassos, tpassos)) # u(x, t) - Solução

# Condições de contorno (Dirichilet)
u[0, :] = t**3       # u(xmin, t) = u(0, t)
u[-1, :] = np.sin(t) # u(xmax, t) = u(2pi, t)

# Solução numérica: diferenças finitas
# Cálculo da temperatura em cada ponto da barra, para todos os tempos
for i in range(1, len(t)):
  for j in range(1, len(x) - 1): # ponta
    u[j, i] = u[j, i-1] + dt*kappa*(u[j-1, i-1] - 2*u[j, i-1] + u[j+1, i-1])/dx**2
```

```python
# Plot da solução numérica
# Distribuição da temperatura na barra para diferentes instantes de tempo (colormap)
import warnings
import matplotlib
warnings.filterwarnings("ignore",category=matplotlib.MatplotlibDeprecationWarning)

from matplotlib import cm
plt.figure(figsize=(8.5, 5))

cmap = cm.rainbow #magma #'gist_rainbow' 'viridis' 'hsv' 'hot_r' 'plasma'
m = cm.ScalarMappable(cmap=cmap) # Mapeia cmap para colorbar
m.set_array(t)                   # Aplica escala de t na colorbar
norm = plt.Normalize(t.min(), t.max())
line_colors = cmap(norm(t))

for i in range(0, tpassos):
    plt.plot(x, u[:,i], label=str(round(i*dt, 2)), color = line_colors[i])
    plt.scatter(x, u[:,i], color = line_colors[i]) # pontos calculados numericamente pela discretização
plt.xlabel('Posição x')
plt.ylabel('Temperatura u(x, t)')
plt.title('Solução numérica equação de calor método de diferenças finitas')
plt.colorbar(m)#, shrink=0.8)
plt.show()
```

```python
# Faça um gráfico da solução aproximada em  u(x,3)  e o ponto  u(2,3)  apontado na solução.
# Dica: calcule a solução aproximada em  u(x,3)  por interplação linear

# Checando discretização de x e t de maneira independente
def pontoxDiscretizacao(pontox, x):
  auxx = 0
  indx = -2
  for i in range(len(x)):
    if (round(x[i], 8) == pontox): #ponto flutuante
      print('Valor x =', pontox, 'Presente na posição índice:', i)
      auxx = 1
      indx = i
      break
  if (auxx == 0):
    print('Valor x =', pontox, 'não está presente na discretização')
  return auxx, indx

def pontotDiscretizacao(pontot, t):
  auxt = 0
  indt = -2
  for i in range(len(t)):
    if (round(t[i], 8) == pontot):  #ponto flutuante
      print('Valor t =', pontot, 'Presente na posição índice:', i)
      auxt = 1
      indt = i
      break
  if (auxt == 0):
    print('Valor t =', pontot, 'não está presente na discretização')
  return auxt, indt

print(x)
print(t)

auxt, indt = pontotDiscretizacao(3.0, t)
```

### Questão 4:  (2,5 pontos)
Considere a função $f(x)=xe^x$.
Faça o gráfico da função $f^{\prime}(x)$ (primeira derivada) e da aproximação de $\hat{f}^{\prime}(x)$ utilizando diferenças finitas centradas no intervalo $[-3,3]$ com $\Delta x=0.1$. Repita o processo para $f^{\prime\prime}(x)$ e $\hat{f}^{\prime\prime}(x)$. O valor dos erros de aproximação (com três casas decimais) $|f^{\prime}(x)-\hat{f}^{\prime}(x)|$ e $|f^{\prime\prime}(x)-\hat{f}^{\prime\prime}(x)|$ no ponto $x=1.5$ são, respectivamente:

a) 0.723 e 0.023<br>
b) 0.756 e 0.022<br>
c) 0.818 e 0.020 (x) <br>
d) 0.812 e 0.023

```python
# Solucao

t = np.linspace(-3, 3, 100)  # domínio da função
def f(x): return x*np.exp(x) # função f(x)

h = 0.1 # delta x
x = 1.5 # ponto de interesse

# Derivada primeira
def df(x): return (x + 1)*np.exp(x)      # solução analítica
def dfapp(x): return (f(x + h) - f(x))/h # aproximação diferenças finitas

# Derivada segunda
def df2(x): return (x + 2)*np.exp(x)                      # solução analítica
def df2app(x): return (f(x + h) - 2*f(x) + f(x - h))/h**2 # aprox diferenças finitas

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))

ax1.plot(t, df(t), label='Solução analítica')
ax1.plot(t, dfapp(t), ls='dashed', label='Aproximação por diferenças finitas')
ax1.scatter(x, dfapp(x), color='magenta', label='Erro x =' + str(x) + ': ' + str(round(abs(df(x) - dfapp(x)), 3)))
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Derivada primeira da função $f(x) = xe^x$')
ax1.legend()

ax2.plot(t, df2(t), label='Solução analítica')
ax2.plot(t, df2app(t), ls='dashed', label='Aproximação por diferenças finitas')
ax2.scatter(x, df2app(x), color='purple', label='Erro x =' + str(x) + ': ' + str(round(abs(df2(x) - df2app(x)), 3)))
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Derivada segunda da função $f(x) = xe^x$')
ax2.legend()

plt.show()
```
